```python
def sigmoid(x):
  import math
  return 1 / (1 + math.exp(-x))
sigmoid(18)
```

This cell defines a Python function `sigmoid` to compute the sigmoid activation, which squashes any real-valued number into a range between 0 and 1. It also provides a demonstration with an input of 18.

```python
def prediction_function(age, affordability):
  weighted_sum = coef[0]*age + coef[1]*affordability + intercept
  return sigmoid(weighted_sum)
```

```python
def log_loss(y_true, y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(i, epsilon) for i in y_predicted]
  y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

This cell defines a `log_loss` function, which is the binary cross-entropy loss function. It's used to quantify the error between the true labels and the predicted probabilities, with `epsilon` to prevent `log(0)` errors.

```python
def sigmoid_numpy(X):
  return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12, 0, 1]))
```

This cell defines a NumPy-optimized sigmoid function `sigmoid_numpy` that can handle array inputs efficiently. It also demonstrates its use with an array of values.

```python
class myNN:
  def __init__(self):
    self.w1 = 1
    self.w2 = 1
    self.bias = 0


  def fit(self, X, y, epochs, loss_threshold):
    self.w1, self.w2, self.bias = self.gradient_descent(X['age'], X['affordability'], y, epochs, loss_threshold)

  def predict(self, X_test):
    weighted_sum = self.w1*X_test['age'] + self.w2*X_test['affordability'] + self.bias
    return sigmoid_numpy(weighted_sum)

  def gradient_descent(self, age, affordability, y_true, epochs, loss_threshold):
    # w1, w2, bias
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age) #  len of samples

    for i in range(epochs):

      weighted_sum = w1 * age + w2 * affordability + bias

      y_predicted = sigmoid_numpy(weighted_sum)

      loss = log_loss(y_true, y_predicted)

      w1d = (1/n)*np.dot(np.transpose(age), (y_predicted-y_true))
      w2d = (1/n)*np.dot(np.transpose(affordability), (y_predicted-y_true))

      bias_d = np.mean(y_predicted-y_true)

      w1 = w1 - rate * w1d
      w2 = w2 - rate * w2d
      bias = bias - rate * bias_d

      if i%50==0:
        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

      if loss<=loss_threshold:
        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        break

    return w1, w2, bias
```

This cell defines a custom neural network class `myNN` from scratch. It includes methods for initialization, fitting the model using gradient descent, and making predictions. The `gradient_descent` method iteratively updates weights and bias to minimize the `log_loss`.

```python
customModel = myNN()
customModel.fit(X_train_scaled, y_train, epochs=500, loss_threshold=0.3378)
```
