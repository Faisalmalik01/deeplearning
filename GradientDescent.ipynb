### **Cell 15: Sigmoid Activation Function**
This cell defines a Python function `sigmoid` that calculates the sigmoid activation for a given input `x`. The sigmoid function squashes any real-valued number into a value between 0 and 1, making it suitable for probabilities in binary classification. It then demonstrates its use with an example.

```python
def sigmoid(x):
  import math
  return 1 / (1 + math.exp(-x))
sigmoid(18)
```

### **Cell 20: Log Loss Function**
This cell defines a `log_loss` function, which is a common loss function for binary classification problems. It incorporates `epsilon` to prevent `log(0)` errors, ensuring numerical stability.

```python
def log_loss(y_true, y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(i, epsilon) for i in y_predicted]
  y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

### **Cell 21: NumPy Sigmoid Function**
This cell defines a `sigmoid_numpy` function that applies the sigmoid activation function to a NumPy array, allowing for element-wise operations. It then demonstrates its use with a sample array.

```python
def sigmoid_numpy(X):
  return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12, 0, 1]))
```

### **Cell 22: Gradient Descent Implementation**
This cell defines the `gradient_descent` function, which implements the logistic regression algorithm from scratch.
- It initializes weights (`w1`, `w2`) and `bias`.
- It iteratively updates these parameters using the learning `rate` to minimize the `log_loss`.
- The gradients (`w1d`, `w2d`, `bias_d`) are calculated based on the difference between predicted and true values.
- It prints the parameters and loss for each epoch. The loop now includes a `loss_threshold` to stop training early if the loss drops below a certain point.

```python
def gradient_descent(age, affordability, y_true, epochs, loss_threshold):
  # w1, w2, bias
  w1 = w2 = 1
  bias = 0
  rate = 0.5
  n = len(age) #  len of samples

  for i in range(epochs):

    weighted_sum = w1 * age + w2 * affordability + bias

    y_predicted = sigmoid_numpy(weighted_sum)

    loss = log_loss(y_true, y_predicted)

    w1d = (1/n)*np.dot(np.transpose(age), (y_predicted-y_true))
    w2d = (1/n)*np.dot(np.transpose(affordability), (y_predicted-y_true))

    bias_d = np.mean(y_predicted-y_true)

    w1 = w1 - rate * w1d
    w2 = w2 - rate * w2d
    bias = bias - rate * bias_d

    print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

    if loss<=loss_threshold:
      break
  return w1, w2, bias
```

```python
gradient_descent(X_train_scaled['age'], X_train_scaled['affordability'], y_train, 1000, 0.3378)
```
