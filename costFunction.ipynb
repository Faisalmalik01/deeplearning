```python
import numpy as np

y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30, 0.7,1,0,0.5])
```

This cell initializes two NumPy arrays: `y_predicted` and `y_true`. These arrays represent predicted and actual values, respectively. They are commonly used in machine learning for calculating loss functions or evaluating model performance.

```python
def mae(y_true, y_predicted):
  total_error = 0
  for yt, yp in zip(y_true, y_predicted):
      total_error += abs(yt - yp)
  print("Total error: ", total_error)
  mae = total_error / len(y_true)
  print("MAE: ", mae)
  return mae
```

This cell defines a Python function called `mae` (Mean Absolute Error). It calculates the average absolute difference between corresponding elements of `y_true` and `y_predicted`. The function then prints the total error and the MAE before returning the MAE value.

```python
mae(y_true, y_predicted)
```

This cell calls the previously defined `mae` function with the `y_true` and `y_predicted` arrays. It demonstrates the usage of the MAE function and the result of the calculation.

This cell calculates the Mean Absolute Error (MAE) using NumPy's vectorized operations. `np.abs(y_predicted - y_true)` computes the absolute differences, and `np.mean()` calculates the average of these differences. This is a more efficient way to calculate MAE compared to the manual loop in the `mae` function.

```python
np.mean(np.abs(y_predicted - y_true))
```

This cell calculates the total absolute error using NumPy. `np.abs(y_predicted - y_true)` computes the absolute differences between the predicted and true values, and `np.sum()` then sums these absolute differences.

```python
np.sum(np.abs(y_predicted - y_true))
```

This cell calculates the Mean Squared Error (MSE) using NumPy. `np.square(y_predicted - y_true)` computes the squared differences between the predicted and true values, and `np.mean()` calculates the average of these squared differences. MSE is another common metric for regression tasks.

```python
np.mean(np.square(y_predicted - y_true))
```

This cell demonstrates the behavior of `np.log` when applied to a very small number. The logarithm of a number close to zero approaches negative infinity, which is relevant for understanding how to handle predictions that might be exactly zero or one in a logarithmic loss function like binary cross-entropy.

```python
np.log([0.000000000000001])
```

This cell defines a small constant `epsilon`. This value is often used in machine learning to prevent division by zero or taking the logarithm of zero, especially in loss functions like log loss (binary cross-entropy) where probabilities can be very close to 0 or 1.

```python
epsilon = 1e-15
```

This cell creates a new list `y_predicted_new` by ensuring that all values in `y_predicted` are at least `epsilon`. This clipping step is crucial for numerical stability when calculating logarithms, as `log(0)` is undefined. It prevents issues if a prediction is exactly 0.

```python
y_predicted_new = [max(i, epsilon) for i in y_predicted]
y_predicted_new
```

This cell further modifies `y_predicted_new` by ensuring that all values are at most `1 - epsilon`. This clipping is also for numerical stability, preventing `log(1 - 1)` (i.e., `log(0)`) if a prediction is exactly 1, which is also undefined. It ensures predictions stay strictly within (0, 1).

```python
y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
y_predicted_new
```

This cell converts the `y_predicted_new` list back into a NumPy array and then applies the natural logarithm (`np.log`) to each element. The output shows the logarithmic values of the clipped predictions.

```python
y_predicted_new = np.array(y_predicted_new)
np.log(y_predicted_new)
```

This cell calculates the binary cross-entropy (or log loss) for a single sample, using the `y_true` and `y_predicted_new` arrays. This formula is commonly used as a loss function for binary classification problems. It measures the performance of a classification model whose output is a probability value between 0 and 1.

```python
-np.mean(y_true*np.log(y_predicted_new) + (1-y_true)*np.log(1-y_predicted_new))
```

This cell defines a function `log_loss` that encapsulates the steps for calculating binary cross-entropy. It includes the clipping of `y_predicted` values using `epsilon` to maintain numerical stability, converts the predictions to a NumPy array, and then applies the log loss formula. This function provides a reusable way to compute the log loss.

```python
def log_loss(y_true, y_predicted):
  y_predicted_new = [max(i, epsilon) for i in y_predicted]
  y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new) + (1-y_true)*np.log(1-y_predicted_new))
```

This cell calls the `log_loss` function with the `y_true` and original `y_predicted` arrays (the clipping is handled within the function). It demonstrates the usage of the `log_loss` function and prints the calculated log loss value.

```python
log_loss(y_true, y_predicted)
```
